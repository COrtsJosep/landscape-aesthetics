```python
import pandas as pd
import os

# data path
directory_path = '/home/ubuntu/landscape-aesthetics/data/processed/wikimedia_commons/dataframes/UK/ns6'
output_file_path = os.path.join(directory_path, 'images_from_uk.csv')

# Define the detection function
def detect_geograph_url(row):
    columns_to_check = ['url', 'image_path', 'metadata_path']
    for column in columns_to_check:
        if pd.notna(row[column]) and 'geograph.org.uk' in row[column]:
            return 1
    return 0

# Create an empty DataFrame to store images from geograph.org.uk
images_from_uk = pd.DataFrame()

# Iterate through all CSV files in a directory
for filename in os.listdir(directory_path):
    if filename.endswith('.csv') and filename != 'images_from_uk.csv':
        file_path = os.path.join(directory_path, filename)
        
        # read scv files
        df = pd.read_csv(file_path)
        
        # Apply the detection function and add a new column
        df['uk_or_not'] = df.apply(detect_geograph_url, axis=1)
        
        # Filter images from geograph.org.uk
        df_uk = df[df['uk_or_not'] == 1]
        
        # add images to images_from_uk DataFrame
        images_from_uk = pd.concat([images_from_uk, df_uk], ignore_index=True)

         # delete images from geograph.org.uk
        #for image_path in df_uk['image_path']:
           # if os.path.exists(image_path):
            #    os.remove(image_path)
        
        # Remove these rows from the original DataFrame
        #df = df[df['uk_or_not'] != 1]
        
        # Save CSV file with new columns
        df.to_csv(file_path, index=False)
        print(f"Processed and updated {filename}")

# Save the images_from_uk DataFrame to a specified file
images_from_uk.to_csv(output_file_path, index=False)
print(f"Saved images from UK to {output_file_path}")
```
