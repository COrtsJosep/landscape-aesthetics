```python
import os

# define dataset path
dataset_dir = '/home/ubuntu/landscape-aesthetics/data/external/ADEChallengeData2016'
print("Dataset directory contents:", os.listdir(dataset_dir))
images_dir = os.path.join(dataset_dir, 'images')
annotations_dir = os.path.join(dataset_dir, 'labels')


print("Images:", os.listdir(images_dir)[:5])  
print("Labels:", os.listdir(annotations_dir)[:5])  

# View the images folder structure
images_train_dir = os.path.join(dataset_dir, 'images/training')
images_val_dir = os.path.join(dataset_dir, 'images/validation')
print("Training images:", os.listdir(images_train_dir)[:5])  
print("Validation images:", os.listdir(images_val_dir)[:5])  

# View the labels folder structure
labels_train_dir = os.path.join(dataset_dir, 'labels/training')
labels_val_dir = os.path.join(dataset_dir, 'labels/validation')
print("Training labels:", os.listdir(labels_train_dir)[:5]) 
print("Validation labels:", os.listdir(labels_val_dir)[:5])  
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Get a list of image and annotation file names


image_files = sorted(f for f in os.listdir(os.path.join(images_dir, 'training')) if f != '.ipynb_checkpoints' and os.path.isfile(os.path.join(images_dir, 'training', f)))
annotation_files = sorted(f for f in os.listdir(os.path.join(annotations_dir, 'training')) if f != '.ipynb_checkpoints' and os.path.isfile(os.path.join(annotations_dir, 'training', f)))
# Ensure that images and annotation files correspond one to one
for img_file, ann_file in zip(image_files, annotation_files):
    print(f"Image file: {img_file}, Annotation file: {ann_file}")

    
    image_path = os.path.join(images_dir, 'training', img_file)
    annotation_path = os.path.join(annotations_dir, 'training', ann_file)

    # loading images
    image = Image.open(image_path)
    
# Instead of opening the annotation file, just use its name
    print(f"Using annotation file name: {ann_file}")

    # view images and annotation file name
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.title(f'Image: {img_file}')

    plt.subplot(1, 2, 2)
    plt.text(0.5, 0.5, ann_file, wrap=True, horizontalalignment='center', verticalalignment='center', fontsize=12)
    plt.title(f'Annotation File: {ann_file}')

    plt.show()

    

    break  
```
